```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', ' Oct', 'Nov', 'Dec']
savings = [0, 100, 150, 200, 300, 500, 300, 350, 500, 600, 750, 100]

plt.plot(months, savings)
plt.xlabel("Months")
plt.ylabel("Savings ($)")
plt.title("Savings Account Results")
plt.show()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label="Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare")
ax.set_ylim(0, 45)
ax.set_title("PyBer Fare by Month")
ax.legend()
# plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# plt.xlabel("Date")
# plt.ylabel("Fare")
# plt.ylim(0, 45)
# plt.title("PyBer Fare by Month")
# plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.scatter(y_axis, x_axis, c="red", marker="x", s=y_axis, label="Chicago")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="skyblue", edgecolors="k", linewidth=2, s=[x*5 for x in y_axis], label="Boston")
ax.invert_yaxis()
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.legend()
```
